# k-Nearest Neighbors

Here we import sklearn's KNeighborsClassifier to classify Gentoo and Chinstrap penguins using the [penguins dataset](https://github.com/mwaskom/seaborn-data/blob/master/penguins.csv "Title").  As always, we start by importing necessary libraries and data, and preprocessing our data:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# settting a nice seaborn theme
sns.set_theme()

# loading the penguin data 
penguins = sns.load_dataset("penguins")

# creating a feature/target matrix 
X = penguins.iloc[154:][["bill_length_mm", "bill_depth_mm"]].values
X = X[~np.isnan(X).any(axis=1)] # removing rows with NaN's
y = np.where(penguins.iloc[154:].species == "Gentoo", 0, 1)
y = np.delete(y, 185)

# creating a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 13)

# instantiating a KNN classifier 
clf = KNeighborsClassifier(n_neighbors = 5)
```

Now that we've imported the necessary libraries/dataset and cleaned our data, we proceed by training our k-nearest neighbors classifier:

```python
clf.fit(X_train, y_train)
```

We then use our k-nearest neighbors classifier to predict if a penguin is Gentoo $(0)$ or Chinstrap $(1)$. We print our model's accuracy as well:

```python
print(f"clf.predict(X_test) = {clf.predict(X_test)} \n")
y_pred = clf.predict(X_test)

print(f"y_pred == y_test = {y_pred == y_test} \n")
print(f"sum(y_pred == y_test)/len(y_test) = {sum(y_pred == y_test)/len(y_test)}")
```

```python
# Calculating the accuracy using the sklearn's score method
clf.score(X_test, y_test)
```

Our model did really well - we only had one misclassification! We plot the decision regions obtained using our classifier below:

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = clf)
plt.xlabel("bill length [mm]", fontsize = 15)
plt.ylabel("bill depth [mm]", fontsize = 15)
plt.title("Decision Boundaries", fontsize = 18)
plt.show()
```

We finish our analysis of the $k$-nearest neighbors algorithm on the penguins dataset by looking at the confusion matrix and the classification report:

```python
from sklearn.metrics import confusion_matrix

# defining the model
clf = KNeighborsClassifier(n_neighbors = 7)

# fitting the model
clf.fit(X_train,y_train)

# making predictions on testing data
y_pred = clf.predict(X_test)

# calculating the confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

# printing the confusion matrix 
print(f"cf_matrix = {cf_matrix} \n")

# viewing the confusion matrix with the seaborn package
plt.figure(figsize = (10, 8))
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues', cbar=False)

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

# tick labels
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

We see from the above confusion matrix that our model correctly idenfified 42 Gentoo penguins (False, or 0 -- the label we assigned to Gentoo penguins) and 20 Chinstrap penguins (True, or 1 -- the label we assigned to Chinstrap penguins). Our only misclassification was an instance of a Gentoo penguin being classified as a Chinstrap penguin. Poor penguin!

```python
from sklearn.metrics import classification_report

# checking performance our model with classification report
print(classification_report(y_test, y_pred))
```

The above classification report shows that our model did really well! Our precision was 1.00 in each case and our weighted average for precision, recall, and f1-scores are all 0.98!

To see how well our $k$-nearest neighbors model does on this data, we compare it first with a logistic regression model and then with a multi-layer perceptron model:

```python
# logistic regression model
from sklearn.linear_model import LogisticRegression

clf2 = LogisticRegression(max_iter = 10_000)
clf2.fit(X_train, y_train)

log_y_pred = clf2.predict(X_test)

plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = clf2)
plt.xlabel("bill length [mm]", fontsize = 15)
plt.ylabel("bill depth [mm]", fontsize = 15)
plt.title("Decision Boundaries", fontsize = 18)
plt.show()
```

The above decision boundary was found by our logistic regression model. Next, we calculate the confusion matrix for this model:

```python
# calculating the confusion matrix
cf_matrix = confusion_matrix(y_test, log_y_pred)

# printing the confusion matrix 
print(f"cf_matrix = {cf_matrix} \n")

# viewing the confusion matrix using the seaborn package
plt.figure(figsize = (10, 8))
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues', cbar=False)

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

# tick labels
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

We see from the above confusion matrix that our logistic regression model correctly idenfified 42 Gentoo penguins and 19 Chinstrap penguins. We misclassfied one Gentoo penguin as a Chinstrap penguin and one Chinstrap penguin as a Gentoo penguin. Hence, it appears as though our k-nearest neighbors classifiers perform (slightly) better than this logistic regression classifier. We finish our analysis of the logistic regression model on this dataset with the following classification report, which does very well overall, but again slightly poorer than the $k$-nearest neighbors model (one can compare this classification report with the above one to see the slight differences in performance):

```python
# checking the performance of our logistic regression model with a classification report
print(classification_report(y_test, log_y_pred))
```

We finish by analyzing how a multi-layer perceptron performs on this dataset:

```python
from sklearn.neural_network import MLPClassifier
clf3 = MLPClassifier(solver='lbfgs', alpha= 0.33,
                   hidden_layer_sizes=(120, 120), random_state=1, max_iter = 50_000)

clf3.fit(X_train, y_train)
net_y_pred = clf3.predict(X_test)

plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = clf3)
plt.xlabel("sepal length [cm]", fontsize = 15)
plt.ylabel("sepal width [cm]", fontsize = 15)
plt.title("Decision Boundaries", fontsize = 18)
plt.show()
```

The above decision regions were found by our multi-layer perceptron model. To see how well it performed as a classifier, we calculate the confustion matrix:

```python
# calculating the confusion matrix
cf_matrix = confusion_matrix(y_test, net_y_pred)

# printing the confusion matrix 
print(f"cf_matrix = {cf_matrix} \n")

# viewing the confusion matrix using the seaborn package
plt.figure(figsize = (10, 8))
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues', cbar=False)

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

# tick labels
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])
plt.show()
```

We see from the above confusion matrix that our multi-layer perceptron model correctly idenfified 40 Gentoo penguins and 19 Chinstrap penguins. We misclassfied 1 Gentoo penguin as a Chinstrap penguin and 3 Chinstrap penguins as Gentoo penguins. Hence this model performs worst among all 3 models on this data!

We finish our analysis of the multi-layer perceptron model on this dataset with the following classification report, which does well overall, but slightly poorer than the $k$-nearest neighbors and the logistic regression model:

```python
# Checking the performance our multi-layer perceptron model with a classification report
print(classification_report(y_test, net_y_pred))
```

This concludes our analysis of the *k*-nearnest neighbors algorithms!
