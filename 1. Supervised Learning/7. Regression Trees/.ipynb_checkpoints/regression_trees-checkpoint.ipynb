```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("abalone.csv") # https://www.kaggle.com/datasets/rodolfomendes/abalone-dataset
```

```python
df
```

```python
X = df[["Length", "Diameter"]].values
y = df["Rings"].values + 1.5 # age of the abalone in years

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], color = "lightseagreen")
plt.xlabel("feature x_0", fontsize = 15)
plt.ylabel("feature x_1", fontsize = 15)
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=3)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 4)

reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

reg1 = DecisionTreeRegressor(max_depth=3, random_state = 4)
reg1.fit(X_train, y_train)
y_pred = reg1.predict(X_test)
print(f"MSE reg1: {mean_squared_error(y_test, y_pred)} \n")


reg2 = DecisionTreeRegressor(max_depth=4, random_state = 4)
reg2.fit(X_train, y_train)
y_pred = reg2.predict(X_test)
print(f"MSE reg2: {mean_squared_error(y_test, y_pred)} \n")
```

```python
from sklearn.tree import plot_tree
from sklearn.tree import export_text

tree_rules = export_text(reg,
                         feature_names = ["x_0", "x_1"])
print(tree_rules, "\n")


plt.figure(figsize=(20, 8))
a = plot_tree(reg,
                   feature_names = ["x_0", "x_1"],
                   class_names = ["red", "blue"],
                   rounded = True,
                   filled = True, 
                   fontsize=14)
plt.show()
```
