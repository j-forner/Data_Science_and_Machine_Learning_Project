# Regression Trees

Here we import the DecisionTreeRegressor from sklearn's tree module to predict the age of abalone based on their length. To start, we import the necessary libraries as well as the [abalone dataset](https://www.kaggle.com/datasets/rodolfomendes/abalone-dataset "Title"):

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("abalone.csv") # importing the data
```

Next, we print the first and last 5 rows of the dataframe to better understand what type of data we have:

```python
df
```

Next, we create a feature vector $X$ that contains length measurements for each of the 4,177 abalone. We store the corresponding real-valued targets $y$ (age in years) as well. Lastly, we include a scatterplot of these abalone instances below:

```python
X = df["Length"].values # feature vector
y = df["Rings"].values + 1.5 # target vector


plt.figure(figsize=(10, 8))
plt.scatter(X[:], y, color = "lightseagreen")
plt.xlabel("Length [mm]", fontsize = 15)
plt.ylabel("Age [years]", fontsize = 15)
plt.title("Abalone Plot")
plt.grid(True)
plt.show()
```

We see in the above plot that there is a positive correlation between the age of abalone and their length. We continue by reshaping our feature vector $X$, and splitting our dataset into training and test sets:

```python
from sklearn.tree import DecisionTreeRegressor

X = X.reshape(-1, 1) # making X a columng vector
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 4)
```

Now that we have training and test data, we instantiate two regression trees "reg1" and "reg2" with a maximum depth of 1 and 5, respectively. We fit each regression tree using the training data, and make predictions using the test data. The mean squared error over all training data for each regression tree is provided below:

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

reg1 = DecisionTreeRegressor(max_depth = 1, random_state = 4)
reg1.fit(X_train, y_train)
y_pred1 = reg1.predict(X_test)
print(f"MSE reg1: {mean_squared_error(y_test, y_pred1)} \n")


reg2 = DecisionTreeRegressor(max_depth = 5, random_state = 4)
reg2.fit(X_train, y_train)
y_pred2 = reg2.predict(X_test)
print(f"MSE reg2: {mean_squared_error(y_test, y_pred2)} \n")
```

We see that the regression tree with a maximum depth of 5 has a lower mean squared error. We conclude our analysis by superimposing the regressors learned from each regression tree onto the first 25 abalone test instances:

```python
x_ax = range(len(y_test[:25]))
plt.figure(figsize = (10, 8))
plt.scatter(x_ax, y_test[:25], color="blue", label="original")
plt.plot(x_ax, y_pred1[:25], color="red", label="predicted (max_depth = 1)")
plt.plot(x_ax, y_pred2[:25], color="green", label="predicted (max_depth = 5)")
plt.legend()
plt.show()
```

The regression tree with a maximum depth of 5 contained more decision rules and as such produced a regressor that better fit the data, as can be seen in the above plot. Although neither is a perfect fit, this exemplifies the idea that a better-trained model will produce better overall predictions to the data. This concludes our analysis of regression trees!
