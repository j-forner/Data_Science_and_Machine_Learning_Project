# *k*-Means Clustering

In this notebook, we import sklearn's KMeans algorithm from its cluster module to cluster houses in California into 6 groups based on geographic location. We use code from this [k-means clustering article](https://www.kaggle.com/code/ryanholbrook/clustering-with-k-means/tutorial "Title"), and start by importing the [california housing dataset](https://www.kaggle.com/datasets/camnugent/california-housing-prices "Title") and different libraries :

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

df = pd.read_csv("housing.csv")
```

Now that we've imported the necessary libraries/data, we take a cursory look at our dataframe:

```python
df
```

We proceed by creating a dataframe $X$ that contains just the median_income, latitude, and longitude columns from the above dataframe. We print the first five rows of this new dataframe below:

```python
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)

X = df.loc[:, ["median_income", "latitude", "longitude"]]
X.head()
```

Now we instantiate a k-means clustering model with $n$ = 6 clusters. We use this model to associate each house with a specific cluster, and add this information as a new column to our dataframe:

```python
# creating cluster feature
kmeans = KMeans(n_clusters=6)
X["Cluster"] = kmeans.fit_predict(X)
X["Cluster"] = X["Cluster"].astype("category")

X.head()
```

We plot each house with respect to its geographic coordinates and with a color associated with whichever of the 6 clusters it belongs to. Looks like California, right?

```python
sns.relplot(
    x="longitude", y="latitude", hue="Cluster", data=X, height=6,
);
```

We see in the above plot the clusters obtained by the KMeans method. Each cluster (by construction) contains points in roughly the same geographic area.

 We finish by adding median_house_value as a column in the dataframe $X$. We include a categorical plot below that displays median house values for each of the 6 clusters:

```python
X["median_house_value"] = df["median_house_value"]
sns.catplot(x="median_house_value", y="Cluster", data=X, kind="boxen", height=6);
```

We see that houses in the blue, red, and brown clisters have the highest median house values. This is due to their clusters being located in more desirable/populated area: San Francisco is located in the area covered by the blue cluster and Los Angeles is located in the area covered by the red and brown clusters.

This concludes our analysis of the *k*-means clustering algorithm on a really interesting dataset!
